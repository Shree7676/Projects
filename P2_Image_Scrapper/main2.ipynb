```python
query="lord+shiva"
google_search=requests.get(f"https://www.google.com/search?q={query}+images&sxsrf=APwXEddJUaLqqKX1_i-oLgw9-5vftsXPpA:1679676924569&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjwmLSThPX9AhUbRmwGHfmGA8gQ_AUoAXoECAEQAw&biw=1536&bih=842&dpr=1.25")
```

```python
beuty=BeautifulSoup(google_search.content,'html.parser')
```

```python
img_tag=beuty.find_all('img')
del img_tag[0]
```

```python
img_tag
```

```python
url_list=list(map(lambda x : x["src"],img_tag))
url_list
```

```python
images=list(map(lambda x : requests.get(x).content,url_list))
images
```

```python
img_dic=list(map(lambda x,y : {"index":x,"image":y},url_list,images))
```

```python
for single_image in images:
    with open(os.path.join(save_dir,f"{query}_{images.index(single_image)}.jpg"),"wb") as f:
        f.write(single_image)
        f.close()
```
